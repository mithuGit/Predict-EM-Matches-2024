```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# Load the data
matches = pd.read_csv("matches.csv", index_col=0)

# Select relevant features
features = ["mp", "w", "d", "l", "gf", "ga", "gd", "pts", "rk"]

# Group by country and calculate mean performance across all years
country_performance = matches.groupby('squad')[features].mean().reset_index()

# Create a composite score
scaler = StandardScaler()
normalized_features = scaler.fit_transform(country_performance[features])

# Define weights for each feature
weights = {
    "mp": 0.05, "w": 0.15, "d": 0.05, "l": -0.1, "gf": 0.1,
    "ga": -0.05, "gd": 0.15, "pts": 0.2, "rk": -0.15
}

# Calculate composite score
composite_scores = np.dot(normalized_features, list(weights.values()))

# Add composite score to the dataframe
country_performance['composite_score'] = composite_scores

# Sort countries by composite score and get top 4
top_4_countries = country_performance.sort_values('composite_score', ascending=False).head(4)

print("Top 4 countries across all years based on composite score:")
for i, (index, row) in enumerate(top_4_countries.iterrows(), 1):
    print(f"{i}. {row['squad']} (Composite Score: {row['composite_score']:.4f})")

print("\nDetailed performance of top 4 countries:")
print(top_4_countries[['squad'] + features].to_string(index=False))

# Calculate how often each country finished in the top 4
top_4_appearances = matches[matches['rk'] <= 4]['squad'].value_counts()

print("\nNumber of top 4 finishes for each country:")
print(top_4_appearances.head(10))  # Show top 10 countries

# Get the top 4 countries based on actual top 4 finishes
actual_top_4 = set(top_4_appearances.head(4).index)

# Compare our predicted top 4 with actual top 4
predicted_top_4 = set(top_4_countries['squad'])

# Calculate accuracy
correct_predictions = len(predicted_top_4.intersection(actual_top_4))
accuracy = correct_predictions / 4  # 4 is the number of top countries we're predicting

print(f"\nAccuracy of the model: {accuracy:.2%}")

print("\nComparison of our predicted top 4 with actual top 4 finishers:")
for country in predicted_top_4:
    appearances = top_4_appearances.get(country, 0)
    in_actual_top_4 = "Yes" if country in actual_top_4 else "No"
    print(f"{country}: {appearances} top 4 finishes, In Actual Top 4: {in_actual_top_4}")

```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler

# Load the data
matches = pd.read_csv("matches.csv", index_col=0)

# Aggregate data for each team across all tournaments
team_stats = matches.groupby('squad').agg({
    'mp': 'sum',
    'w': 'sum',
    'd': 'sum',
    'l': 'sum',
    'gf': 'sum',
    'ga': 'sum',
    'gd': 'sum',
    'pts': 'sum',
    'year': 'count'  # Number of tournaments participated
}).reset_index()

# Calculate additional features
team_stats['avg_pts_per_match'] = team_stats['pts'] / team_stats['mp']
team_stats['avg_gd_per_match'] = team_stats['gd'] / team_stats['mp']
team_stats['win_rate'] = team_stats['w'] / team_stats['mp']

# Sort teams by average points per match and select top 4
top_4_teams = team_stats.nlargest(4, 'avg_pts_per_match')['squad'].tolist()

# Create target variable
team_stats['Top_4'] = team_stats['squad'].isin(top_4_teams)

print(team_stats)

# Select features for the model
features = ['mp', 'w', 'd', 'l', 'gf', 'ga', 'gd', 'pts', 'year', 'avg_pts_per_match', 'avg_gd_per_match', 'win_rate']
X = team_stats[features]
y = team_stats['Top_4']

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42, stratify=y)

# Train the model
clf = RandomForestClassifier(random_state=42, class_weight='balanced')
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(report)

# Feature importance
feature_importance = pd.DataFrame({'feature': features, 'importance': clf.feature_importances_})
feature_importance = feature_importance.sort_values('importance', ascending=False)
print("\nFeature Importance:")
print(feature_importance)

# Print top 4 teams
print("\nTop 4 Teams based on average points per match:")
print(top_4_teams)

# Predict probabilities for all teams
team_stats['top_4_probability'] = clf.predict_proba(X_scaled)[:, 1]

# Print top 10 teams by model prediction
print("\nTop 10 Teams based on model prediction:")
print(team_stats.nlargest(10, 'top_4_probability')[['squad', 'top_4_probability']])

# Analyze misclassifications
misclassified = team_stats[clf.predict(X_scaled) != team_stats['Top_4']]
print("\nMisclassified Teams:")
print(misclassified[['squad', 'Top_4', 'top_4_probability']])
```
